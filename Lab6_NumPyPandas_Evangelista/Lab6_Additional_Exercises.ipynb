```python
import numpy as np
import pandas as pd

STUDENT_NAME = "Gian A. Evangelista"
STUDENT_ID = 250295

languages = sorted(['C#', 'Flutter', 'Java', 'JavaScript', 'Matlab', 'PhP', 'Python', 'React', 'Swift', 'TypeScript'])
assigned_index = (STUDENT_ID % 1000) % 10  # 756 % 10 = 6
assigned_lang = languages[assigned_index]   # Python
second_lang = languages[assigned_index + 1] # React

print(f"LABORATORY REPORT: {STUDENT_NAME} ({STUDENT_ID})")
print(f"Assigned Languages: {assigned_lang} (Primary) and {second_lang} (Secondary)\n")
```

```python
assigned_lang = "Python" # Change "PhP" to "Python" to match the test file

# ADDITIONAL EXERCISE 1: Programming Language Growth Analysis
import numpy as np
import pandas as pd


print("ADDITIONAL EXERCISE 1 – Programming Language Growth Analysis")

df = pd.read_csv('Most-Popular-Programming-Languages.csv')
df['Month'] = pd.to_datetime(df['Month'])

# Find the correct column name
pop_col = None
for col in df.columns:
    if assigned_lang.lower() in col.lower():
        pop_col = col
        break

if pop_col is None:
    raise KeyError(f"Could not find column for '{assigned_lang}'. Available: {list(df.columns)}")

print(f"Assigned Language: {assigned_lang}")
print(f"Column used: {pop_col}\n")

df_analysis = df[['Month', pop_col]].copy()
df_analysis.columns = ['Month', 'Popularity']

df_analysis['Growth_Rate'] = df_analysis['Popularity'].pct_change() * 100
df_analysis['Moving_Avg']  = df_analysis['Popularity'].rolling(window=12).mean()
df_analysis['Moving_STD']  = df_analysis['Popularity'].rolling(window=12).std()

conditions = [
    df_analysis['Growth_Rate'] > 5.0,
    df_analysis['Growth_Rate'] < -5.0
]
choices = ['Growth', 'Decline']
df_analysis['Phase'] = np.select(conditions, choices, default='Stable')

print("--- DataFrame Output (first 10 rows) ---")
print(df_analysis.head(10).to_string(index=False))

print("\n--- Statistical Summary ---")
print(df_analysis['Popularity'].describe())

print("\n--- Growth Phase Counts ---")
print(df_analysis['Phase'].value_counts())

initial_pop    = df_analysis['Popularity'].iloc[0]
final_pop      = df_analysis['Popularity'].iloc[-1]
overall_growth = ((final_pop - initial_pop) / initial_pop) * 100
phase_counts   = df_analysis['Phase'].value_counts()

print("\n--- Key Performance Metrics ---")
print(f"Initial Popularity  : {initial_pop:.2f}%")
print(f"Final Popularity    : {final_pop:.2f}%")
print(f"Overall Growth      : {overall_growth:.2f}%")
print(f"Growth Periods      : {phase_counts.get('Growth', 0)}")
print(f"Decline Periods     : {phase_counts.get('Decline', 0)}")
print(f"Stable Periods      : {phase_counts.get('Stable', 0)}")
```

```python
# ADDITIONAL EXERCISE 2: 
import numpy as np
import pandas as pd

print("ADDITIONAL EXERCISE 2 – Lifecycle Classification")

df2 = pd.read_csv('Most-Popular-Programming-Languages.csv')
df2['Month'] = pd.to_datetime(df2['Month'])

pop_col2 = None
for col in df2.columns:
    if assigned_lang.lower() in col.lower():
        pop_col2 = col
        break

print(f"Assigned Language: {assigned_lang}")
print(f"Column used: {pop_col2}\n")

df_life = df2[['Month', pop_col2]].copy()
df_life.columns = ['Month', 'Popularity']

df_life['Growth_Rate'] = df_life['Popularity'].pct_change() * 100
df_life['Moving_Avg']  = df_life['Popularity'].rolling(window=6).mean()
df_life['Moving_STD']  = df_life['Popularity'].rolling(window=6).std()

gr_mean = np.mean(df_life['Growth_Rate'].dropna())
gr_std  = np.std(df_life['Growth_Rate'].dropna())

print(f"Growth Rate Mean : {gr_mean:.4f}")
print(f"Growth Rate STD  : {gr_std:.4f}\n")

gr = df_life['Growth_Rate']

lc_conditions = [
    (gr > 0) & (gr < gr_mean),
    gr > gr_mean,
    np.abs(gr) <= 1,
    (gr < 0) & (gr < (-1 * gr_std))
]
lc_choices = ['Introduction', 'Growth', 'Maturity', 'Decline']
df_life['Lifecycle_Phase'] = np.select(lc_conditions, lc_choices, default='Stable')

print("--- Lifecycle Classification Table (first 15 rows) ---")
print(df_life.head(15).to_string(index=False))

phase_counts2 = df_life['Lifecycle_Phase'].value_counts()
phase_pct2    = df_life['Lifecycle_Phase'].value_counts(normalize=True) * 100

print("\n--- Lifecycle Phase Summary ---")
summary2 = pd.DataFrame({'Count': phase_counts2, 'Percentage (%)': phase_pct2.round(2)})
print(summary2)

dominant_phase = phase_counts2.idxmax()
print(f"\nDominant Stage: {assigned_lang} is currently in the '{dominant_phase}' phase.")

print("\n--- Interpretive Insights ---")
overall_growth2 = ((df_life['Popularity'].iloc[-1] - df_life['Popularity'].iloc[0]) / df_life['Popularity'].iloc[0]) * 100
print(f"Overall growth from first to last record: {overall_growth2:.2f}%")
print(f"{assigned_lang} shows {phase_counts2.get('Growth', 0)} growth periods and {phase_counts2.get('Decline', 0)} decline periods across the 20-year span.")
```

```python
assigned_lang = "Python"
second_lang = "JavaScript" # Change this to a different language

# ADDITIONAL EXERCISE 3: 
import numpy as np
import pandas as pd

print("ADDITIONAL EXERCISE 3 – Comparative Analysis")
print(f"  {assigned_lang} vs {second_lang}")

df3 = pd.read_csv('Most-Popular-Programming-Languages.csv')
df3['Month'] = pd.to_datetime(df3['Month'])

col_a, col_b = None, None
for col in df3.columns:
    if assigned_lang.lower() in col.lower():
        col_a = col
    if second_lang.lower() in col.lower():
        col_b = col

if col_a is None or col_b is None:
    raise KeyError(f"Missing columns. Found: col_a={col_a}, col_b={col_b}. Available: {list(df3.columns)}")

print(f"Language A: {assigned_lang} → Column: {col_a}")
print(f"Language B: {second_lang} → Column: {col_b}\n")

df_comp = df3[['Month', col_a, col_b]].copy()
df_comp.columns = ['Month', assigned_lang, second_lang]

lang_a = df_comp[assigned_lang].values
lang_b = df_comp[second_lang].values

mean_a   = np.mean(lang_a)
mean_b   = np.mean(lang_b)
median_a = np.median(lang_a)
median_b = np.median(lang_b)
std_a    = np.std(lang_a)
std_b    = np.std(lang_b)

correlation     = np.corrcoef(lang_a, lang_b)[0, 1]
count_a_dominant = np.sum(lang_a > lang_b)
total_months    = len(lang_a)
dominance_ratio = (count_a_dominant / total_months) * 100
rpi             = mean_a / mean_b if mean_b != 0 else np.nan

prev_a_gt_b    = lang_a[:-1] <= lang_b[:-1]
curr_a_gt_b    = lang_a[1:]  >  lang_b[1:]
crossover_mask = prev_a_gt_b & curr_a_gt_b
# First, filter the months to match the size of your crossover_mask
months_subset = df_comp['Month'].iloc[1:]

# Then apply the mask to that subset
crossover_months = months_subset[crossover_mask].values

print("--- Comparative Statistics ---")
stats_df = pd.DataFrame({
    'Metric': ['Mean', 'Median', 'Std Dev', 'Correlation', 'Dominance Ratio (%)', 'RPI'],
    assigned_lang: [mean_a, median_a, std_a, correlation, dominance_ratio, rpi],
    second_lang:   [mean_b, median_b, std_b, correlation, 100 - dominance_ratio, 1/rpi if rpi != 0 else np.nan]
})
print(stats_df.to_string(index=False))

print("\n--- Summary Row ---")
summary_row = pd.DataFrame({
    'Mean_A':          [round(mean_a, 4)],
    'Mean_B':          [round(mean_b, 4)],
    'Std_A':           [round(std_a, 4)],
    'Std_B':           [round(std_b, 4)],
    'Correlation':     [round(correlation, 4)],
    'Dominance_Ratio': [round(dominance_ratio, 2)],
    'RPI':             [round(rpi, 4)]
})
print(summary_row.to_string(index=False))

print(f"\n--- Crossover Points ({assigned_lang} surpassing {second_lang}) ---")
if len(crossover_months) > 0:
    for m in crossover_months:
        print(f"  {m}")
else:
    print("  No crossover points found — one language consistently dominates.")

print("\n--- Interpretation ---")
winner = assigned_lang if dominance_ratio > 50 else second_lang
print(f"  {winner} dominates in popularity for {max(dominance_ratio, 100 - dominance_ratio):.1f}% of recorded months.")
corr_desc = 'strongly positively' if correlation > 0.7 else ('weakly' if abs(correlation) < 0.3 else 'moderately')
print(f"  {assigned_lang} and {second_lang} are {corr_desc} correlated (r = {correlation:.4f}).")
more_stable = assigned_lang if std_a < std_b else second_lang
print(f"  {more_stable} shows more consistent popularity over time (lower STD).")
```
