```python
# Procedure 1: Writing to a File

# Step 1: Import Path from pathlib
from pathlib import Path

# Step 2: Define a custom folder in your Documents directory
output_dir = Path.home() / "Documents" / "Evangelista_Activity_5"
output_dir.mkdir(exist_ok=True) # Creates folder if missing

# Step 3: Define the file path
file_path = output_dir / "Act5_example.txt"

# Step 4: Write text to the file
with open(file_path, "w", encoding="utf-8") as file:
    file.write("Hello, Welcome to Python Programming!\n")
    file.write("File saved safely with pathlib.\n")
    file.write("Python makes file handling easy!\n")

# Step 5: Confirm that the file is saved
print(f"File saved to: {file_path.resolve()}")
```

```python
# Updated Procedure 2: Reading from a file with filters and word counts

# Step 2: Read line-by-line
with open(file_path, "r", encoding="utf-8") as file:
    print("--- Try It Results ---")
    for line_number, line in enumerate (file, 1):
        clean_line = line.strip()           #remove extra spaces/newlines
        words = clean_line.split()          #split into words
        word_count = len(words)             #count words
        if "Python" in clean_line:          #filter lines containing "Python"
            print(f"Line {line_number} (Words: {word_count}): {clean_line}")

```

```python
# Procedure 3: Appending Data

# Step 1: Open file in append mode
with open(file_path, "a", encoding="utf-8") as file:
    file.write("\nThis line was added!")
    
    # Try It: Append multiple lines using a list and file.writelines().
    extra_lines = [
        "\nAdding more content via list.",
        "\nFile handling in Python is powerful!"
    ]
    file.writelines (extra_lines)

    # Try It: Ask the user for input and append that to the file.
    user_note = input("Enter a sentence to add to your file: ")
    file.write(f"\nUser added: {user_note}")
    
# Step 2: Confirm that data was added
print("Data appended successfully.")
```

```python
# Procedure 4: Safe File Operations with Backup

# Step 1: Import additional libraries
from pathlib import Path
import shutil
from datetime import datetime

student_id = "2025-0295"

# Step 2: Define working directory
backup_dir = Path.home() / "Documents" / "Evangelista_Activity_5" / f"backup_{student_id}"
backup_dir.mkdir(exist_ok=True, parents=True)

# Search for files that start with the filename and contain 'backup'
def write_with_backup(filename_str, content):
    # Convert the string filename to a Path object to use .stem and .suffix
    filename = Path(filename_str)
    # We must ensure we are looking in the correct output directory
    full_path = Path.home() / "Documents" / "Evangelista_Activity_5" / filename
    
    # Create a unique backup name using a timestamp
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    backup_name = f"{full_path.stem}_Evangelista_backup_{timestamp}{full_path.suffix}"
    backup_path = backup_dir / backup_name

    # Step 1: Create a backup of the existing file
    if full_path.exists():
        shutil.copy(full_path, backup_path)
        print(f"Backup created: {backup_path.name}")

    # Step 2: Write new content to the original file
    with open(full_path, "w", encoding="utf-8") as file:
        file.write(content)
    print(f"File '{full_path.name}' updated successfully.")

# Step 3: Define function to list specific backups
def list_specific_backups(filename_str):
    filename = Path(filename_str)
    print(f"\n--- Backups for {filename.name} ---")
    # Search pattern for glob
    search_pattern = f"{filename.stem}_backup*"
    for backup in backup_dir.glob(search_pattern):
        print("-", backup.name)

# --- Execution Section ---
print("=== File Operations Demo ===")

# 1. Initial creation
print("\n1. Creating new file:")
write_with_backup("demo.txt", "Initial content")

# 2. Update (This will trigger the backup)
print("\n2. Updating file (with backup):")
write_with_backup("demo.txt", "Updated content")

# 3. Listing backups
list_specific_backups("demo.txt")
```

```python
# Procedure 5: Menu-Driven File Manager
# Step 1: Import required modules
from pathlib import Path
from datetime import datetime
import shutil

# Step 2: Define your custom working directory in Documents 
backup_dir = Path.home() / "Documents" / "Evangelista_Activity_5" 
backup_dir.mkdir(exist_ok=True)  # Create folder if it doesn’t exist 

# Step 3: Define the menu-driven file manager 
def file_manager(): 
    file_name = input("Enter filename (e.g., notes.txt): ") 
    file_path = backup_dir / file_name 

    while True: 
        print("\n--- MENU ---") 
        print("1. Write to file") 
        print("2. Append to file") 
        print("3. Read file") 
        print("4. Backup files")
        print("5. List all backup files") 
        print("6. Exit") 

        choice = input("Choose an option (1-6): ") 

        if choice == "1": 
            content = input("Enter content to write:\n") 
            with open(file_path, "w", encoding="utf-8") as f: 
               f.write(content) 
            print("File written successfully.") 

        elif choice == "2": 
            more = input("Enter content to append:\n") 
            with open(file_path, "a", encoding="utf-8") as f: 
                f.write("\n" + more) 
            print("Content appended.") 

        elif choice == "3": 
            if file_path.exists(): 
                with open(file_path, "r", encoding="utf-8") as f: 
                    print("\nFile Content:\n", f.read())
            else: 
                print("File not found.") 

        elif choice == "4": 
            if file_path.exists(): 
                timestamp = datetime.now().strftime("%Y%m%d_%H%M%S") 
                backup_file = file_path.with_name( 
                    f"{file_path.stem}_backup_{timestamp}{file_path.suffix}" 
                ) 
                shutil.copy2(file_path, backup_file) 
                print(f"Backup created: {backup_file.name}") 
            else: 
                print("Cannot backup. File does not exist.") 

        elif choice == "5": 
            # Try It: Add a menu option to list all backup files
            print(f"\n--- All Backups in {backup_dir.name} ---")
            for backup in backup_dir.glob("backup"):
                print("-", backup.name)
        
        elif choice == "6": 
            print("Exiting the file manager.") 
            break 

        else: 
            print("Invalid choice. Please try again.") 
            
# Step 4: Run the menu-driven tool 
if __name__ == "_main_": 
    file_manager()
```
